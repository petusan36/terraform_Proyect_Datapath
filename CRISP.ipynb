# Top 10 jugadores más costosos según rango de edad y posición:	gráfica

```python
df_players[["edad"]].info()
```

```python
df_players.loc[(df_players["edad"]>= 16) & (df_players["edad"] <= 22), "rango_edad"] = "16-22"
df_players.loc[(df_players["edad"]>= 23) & (df_players["edad"] <= 25), "rango_edad"] = "23-25"
df_players.loc[(df_players["edad"]>= 26) & (df_players["edad"] <= 29), "rango_edad"] = "26-29"
df_players.loc[(df_players["edad"]>= 30 ), "rango_edad"] = "30+"

top_10_costos=df_players[["id","nombre_largo","rango_edad","posiciones_jugador","salario_euros","valor_euros"]]
top_10_costos["rank_salario"]=top_10_costos.groupby(["rango_edad","posiciones_jugador"])["salario_euros"].rank(method="dense", ascending=False)
top_10_costos["rank_valor"]=top_10_costos.groupby(["rango_edad","posiciones_jugador"])["valor_euros"].rank(method="dense", ascending=False)

top_10_costos_salario=top_10_costos[["id","nombre_largo","rango_edad","posiciones_jugador","salario_euros","rank_salario"]][top_10_costos["rank_salario"]<=10].sort_values(by=["rank_salario", "posiciones_jugador"], ascending= True)
top_10_costos_valor=top_10_costos[["id","nombre_largo","rango_edad","posiciones_jugador","valor_euros","rank_valor"]][top_10_costos["rank_salario"]<=10].sort_values(by=["rank_valor","posiciones_jugador"], ascending= True)
```

# Top 5 jugadores más efectivos según la posición:	tabla resumen 

> Criterios de efectividad: 

Para esto se van a sumar los siguientes campos que daran un indicador de calidad, obteniendo de allí los totales para su clasificación

* movimientos_calidad

```python
columnas_criterio= ["movimientos_calidad","ritmo_jugador","disparo",\
                   "pase","dribbling","defensa","fisico","cruce_ataque","finalidad_ataque","exactitud_ataque_cabeza", \
                   "ataque_pase_corto", "disparo_jugador","habilidad_dribling","habilidad_curva","habilidad_fk_exactitud",\
                   "habilidad_pase_largo","habilidad_control_balon","aceleracion_movimiento","movimiento_velocididad_aceleracion",\
                   "agilidad_movimiento","reacciones_movimiento","balance_movimiento","poder_disparo","poder_salto","poder_aguante",\
                   "poder_fortaleza","poder_disparos_largo"]

top_5_efectivos=df_players[["id","nombre_largo","posiciones_jugador","movimientos_calidad","ritmo_jugador","disparo",\
                   "pase","dribbling","defensa","fisico","cruce_ataque","finalidad_ataque","exactitud_ataque_cabeza", \
                   "ataque_pase_corto", "disparo_jugador","habilidad_dribling","habilidad_curva","habilidad_fk_exactitud",\
                   "habilidad_pase_largo","habilidad_control_balon","aceleracion_movimiento","movimiento_velocididad_aceleracion",\
                   "agilidad_movimiento","reacciones_movimiento","balance_movimiento","poder_disparo","poder_salto","poder_aguante",\
                   "poder_fortaleza","poder_disparos_largo"]]

col_suma_criterio=top_5_efectivos[columnas_criterio].fillna(0)
top_5_efectivos["suma_criterios"]= col_suma_criterio.sum(axis=1)

top_5_efectivos["rank_efectivos"]=top_5_efectivos.groupby(["posiciones_jugador"])[["suma_criterios"]].rank(method="dense", ascending=False)
top_5_efectivos_posicion=top_5_efectivos[["id","nombre_largo","posiciones_jugador","rank_efectivos"]][top_5_efectivos["rank_efectivos"]<=5].sort_values(by=["rank_efectivos"], ascending=True)

with pd.option_context('display.max_colwidth', None,
                       'display.max_columns', None,
                       'display.max_rows', None,
                       'display.max_seq_items', None):
    display(top_5_efectivos_posicion.head(10))
```
